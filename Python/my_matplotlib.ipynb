```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

# 1. Functional

```python
plt.plot(x,y,'b-.')
plt.xlabel('Ku to nhu phich')
plt.ylabel('Ku ta to ghe')
plt.title('Hehe haha')
```

```python
# many plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# 2. Object oriented

## create axes manually

```python
fig=plt.figure()

axes1=fig.add_axes([.1,.1,.7,.8]) # from left of canvas (%), from bottom of canvas, width, height
axes2=fig.add_axes([.2,.3,.4,.5])

axes1.plot(x,y,'b-.')
axes1.set_xlabel('Ku')
axes1.set_ylabel('Chim')
axes1.set_title('Outer')

axes2.plot(y,x,'m:')
axes2.set_xlabel('To')
axes2.set_ylabel('Be')
axes2.set_title('Inner')
```

# 3. Subplots on OO

```python
fig,axes = plt.subplots(nrows=1,ncols=2) # automatically call add axes, quite reasonable scale, but not always good!
plt.tight_layout() # fix overlapping errors...
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) 

# not that 'axes' here is an array of axes, so we can loop or index
for axe in axes:
    axe.plot(x,y)
```

#### nothing here, must be in the same cell of 'fig,axes' to be drawed

```python
axes[0].plot(x,y,'y:')
axes[1].plot(y,x,'b--')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'y:')
axes[1].plot(y,x,'b--')
```

# 6. Legend

- ## Used to remember names of plots on the same figure and the same axe

```python
fig=plt.figure()

ax=fig.add_axes([.1,.1,.8,.9])

ax.plot(x,x**2,label='squared')
ax.plot(x,x**3,label='cubed')
ax.legend()
```

```python
# can add 'loc' inside legend to specify position
# loc=0 or 1,2,...10 or loc=(a,b) is the coordinate of low left corner
```

# 7. Colors, customization of lines, markers

```python
fig=plt.figure()

ax=fig.add_axes([.1,.1,.8,.9])
ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=.7,linestyle='-.',marker='*',markersize=20,
       markerfacecolor='b',markeredgewidth=2, markeredgecolor='black') # RGB hex code

# appha  is transparency
# linewidth  = lw
# linestyle = ls
```

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,.8,.9])
ax.plot(y,marker='o')
```

# 8. Range

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,.8,.9])
ax.plot(x,y,lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,5])
```
